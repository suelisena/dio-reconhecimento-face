```python
import face_recognition
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt # Para exibir imagens no Jupyter
```

```python
# --- Parte 1: Codificando os Rostos Conhecidos ---

# Caminho para a pasta com as imagens de referência dos personagens
# Certifique-se que 'faces_friends' está no mesmo diretório do seu notebook
known_faces_dir = r'C:\Users\sueli\anaconda_projects\dio-reconhecimento_face\face-friends'

known_face_encodings = []
known_face_names = []

# Carregar e codificar os rostos conhecidos
print("Carregando e codificando rostos conhecidos...")
for filename in os.listdir(known_faces_dir):
    if filename.endswith((".jpg", ".png", ".jpeg")):
        path = os.path.join(known_faces_dir, filename)
        image = face_recognition.load_image_file(path)

        # Encontra as codificações (embeddings) de todos os rostos na imagem
        # Geralmente, haverá apenas um rosto por imagem de referência
        face_encodings = face_recognition.face_encodings(image)

        if len(face_encodings) > 0:
            # Pega a primeira (e geralmente única) codificação de rosto
            known_face_encodings.append(face_encodings[0])

            # O nome do rosto é o nome do arquivo (sem a extensão)
            known_face_names.append(os.path.splitext(filename)[0])
            print(f"Rosto de '{known_face_names[-1]}' codificado.")
        else:
            print(f"Nenhum rosto encontrado na imagem: {filename}")

print(f"Total de rostos conhecidos carregados: {len(known_face_names)}")


# --- Parte 2: Reconhecendo Rostos na Imagem da Série Friends ---

# Caminho para a imagem da série Friends a ser analisada
# Certifique-se que 'friends_scene.jpg' está no mesmo diretório do seu notebook
friends_image_path = r'C:\Users\sueli\anaconda_projects\dio-reconhecimento_face\face-friends\todos\friends_scene.jpeg'

# Carregar a imagem da série Friends
print(f"\nAnalisando imagem: {friends_image_path}")
image_to_recognize = face_recognition.load_image_file(friends_image_path)
# Converter a imagem para um formato que o OpenCV possa usar (BGR)
# e também para exibir com Matplotlib (RGB)
image_display = cv2.cvtColor(image_to_recognize, cv2.COLOR_RGB2BGR) # Para desenhar
image_for_plot = image_to_recognize.copy() # Para exibir no Matplotlib

# Encontrar todos os rostos e suas codificações na imagem da série
face_locations = face_recognition.face_locations(image_to_recognize)
face_encodings = face_recognition.face_encodings(image_to_recognize, face_locations)

print(f"Total de rostos detectados na imagem: {len(face_locations)}")

# Processar cada rosto encontrado
for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
    name = "Desconhecido"
    
    # Comparar o rosto encontrado com os rostos conhecidos
    matches = face_recognition.compare_faces(known_face_encodings, face_encoding)

    # Encontrar a melhor correspondência usando a distância euclidiana
    face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
    best_match_index = np.argmin(face_distances)

    # Se a melhor correspondência for um 'match' e estiver dentro de um limite aceitável
    # O limite padrão de 'compare_faces' é 0.6. distâncias menores são melhores.
    # Você pode ajustar esse limite (e.g., < 0.5) para ser mais rigoroso.
    if matches[best_match_index] and face_distances[best_match_index] < 0.6: # Exemplo de limite
        name = known_face_names[best_match_index]
        print(f"Rosto encontrado: {name} (confiança: {1 - face_distances[best_match_index]:.2f})")
    else:
        print(f"Rosto desconhecido encontrado (distância: {face_distances[best_match_index]:.2f}).")

    # Desenhar um retângulo e o nome no rosto
    cv2.rectangle(image_display, (left, top), (right, bottom), (0, 255, 0), 2)

    # Definir o tipo de fonte (pode testar outros como FONT_HERSHEY_SIMPLEX)
    font = cv2.FONT_HERSHEY_SIMPLEX # Geralmente mais suave que DUPLEX
     # Ajustar o tamanho e a espessura do texto
    font_scale = 0.4 # Aumentado de 0.3 para 0.7 (ajuste conforme necessário)
    font_thickness = 1 # Aumentado de 1 para 2 (ajuste conforme necessário)

    # Calcular a posição do texto para que fique acima do retângulo ou dentro
    # Uma boa prática é colocar um pouco acima do top do retângulo
    text_x = left + 4
    text_y = top - 10 # 10 pixels acima da borda superior do retângulo

    # Para garantir que o texto não saia da imagem se o rosto estiver muito no topo
    if text_y < 10: # Se estiver muito perto da borda superior da imagem
        text_y = top + 20 # Coloca dentro do retângulo, um pouco abaixo do topo

    # Desenhar o texto com as novas configurações
    # (255, 255, 255) é a cor branca para o texto
    cv2.putText(image_display, name, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness)

# --- Parte 3: Exibir a Imagem no Jupyter Notebook ---

# O OpenCV usa BGR, mas o Matplotlib usa RGB. Precisamos converter de volta.
image_display_rgb = cv2.cvtColor(image_display, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8)) # Ajuste o tamanho da figura conforme necessário
plt.imshow(image_display_rgb)
plt.axis('off') # Remove os eixos
plt.show()

print("\nProcesso concluído.")
```
